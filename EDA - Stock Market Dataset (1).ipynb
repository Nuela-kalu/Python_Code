```python
#Importing libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing dataset

df = pd.read_excel("Stock Market dataset.xlsx")
```

```python
#Reviewing data structure

#Display the first 5 rows of the DataFrame
print(df.head(5))
```

```python
#Display the dimensions of the DataFrame
print("Shape of DataFrame:", df.shape)
```

```python
#List all columns in the DataFrame
print("Columns in DataFrame:", df.columns)
```

```python
#Check for missing values
print("Missing Values:\n", df.isnull().sum())
```

```python
#Display summary statistics for numerical variables
print("Summary Statistics:\n", df.describe())
```

```python
#Visualize Data

#Histogram of a Ethereum Price
plt.hist(df['Ethereum Price'])
plt.xlabel('Ethereum Price')
plt.ylabel('Frequency')
plt.title('Histogram of Ethereum Price')
plt.show()
```

```python
#Relationships between Apple and Tesla

#Scatter plot of Apple and Tesla
plt.scatter(df['Apple Price'], df['Tesla Price'])
plt.xlabel('Apple Price')
plt.ylabel('Tesla Price')
plt.title('Scatter Plot of Apple and Tesla Prices')
plt.show()
```
